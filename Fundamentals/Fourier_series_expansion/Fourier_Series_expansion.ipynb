```python
# -*- coding: utf-8 -*-
"""
Fourier series expansion - analytical
"""
import numpy as np
import matplotlib.pyplot as plt
import math
A=1
T=2
 
x=np.arange(-T,2*T,0.01)
y=np.zeros(shape=(x.shape[0],))
for i in range(x.shape[0]):
    if (x[i] < (-T/2)) or ((x[i] > 0) and (x[i] < T/2)) or ((x[i] > T) and (x[i] < 3*T/2)):
        y[i] = 1
     
 
fig= plt.figure(figsize=(8,4))
plt.xlim(-2.5, 4.5)
plt.ylim(-0.2, 1.2)
plt.axis('equal')
plt.ylabel('f(t)')
plt.xlabel('time - t')
plt.plot(x,y,'b', linewidth=2.5, label='Periodic function')
plt.savefig('periodic_fuction.eps')
 
 
kmax_list = [1, 3, 5, 10, 20, 100]
w0 = 2 * math.pi / T
fig, axs = plt.subplots(2, 3, figsize=(18, 8))
axs = axs.flatten()
 
for idx, k_max in enumerate(kmax_list):
    yapprox = []
    for i in range(x.shape[0]):
        s = 0
        for k in np.arange(1, k_max+1):
            s += A * ((math.sin(k*math.pi/2)) / (k*math.pi/2)) * math.cos(k*w0*x[i] - k*math.pi/2)
        yapprox.append(s + 0.5*A)
    axs[idx].plot(x, y, 'b', linewidth=2.0, label='Periodic function')
    axs[idx].plot(x, yapprox, 'r', linewidth=2.0, label=f'k_max={k_max}')
    axs[idx].set_xlim(-2.5, 4.5)
    axs[idx].set_ylim(-0.2, 1.2)
    axs[idx].set_title(f'Fourier Approximation (k_max={k_max})')
    axs[idx].set_xlabel('time - t')
    axs[idx].set_ylabel('f(t)')
    axs[idx].legend()

plt.tight_layout()
plt.show()
```
