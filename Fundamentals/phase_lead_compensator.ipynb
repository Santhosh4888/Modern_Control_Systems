```python
# phase_lead.py

import matplotlib.pyplot as plt
import control

# Define the transfer function W = 10 / (s^2 + s)
num = [10]
den = [1, 1, 0]
W = control.TransferFunction(num, den)

# Bode plot
plt.figure(1)
control.bode_plot(W, dB=True, Hz=False, grid=True)
plt.show()

# Closed-loop feedback system (unity feedback)
Wfeedback = control.feedback(W, 1)

# Step response
plt.figure(2)
t, y = control.step_response(Wfeedback)
plt.plot(t, y)
plt.grid(visible=True)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Output')
```
